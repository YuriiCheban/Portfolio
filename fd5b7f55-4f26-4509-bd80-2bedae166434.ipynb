### –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∏ –¥–æ–±–∞–≤—å—Ç–µ –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã

- —Ü–µ–Ω–∞ –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ - price_m2;
- –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è (0 ‚Äî –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 1 ‚Äî –≤—Ç–æ—Ä–Ω–∏–∫ –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ) - num_day;
- –º–µ—Å—è—Ü –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è - num_month;
- –≥–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è - num_year;
- —Ç–∏–ø —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã (–∑–Ω–∞—á–µ–Ω–∏—è ‚Äî ¬´–ø–µ—Ä–≤—ã–π¬ª, ¬´–ø–æ—Å–ª–µ–¥–Ω–∏–π¬ª, ¬´–¥—Ä—É–≥–æ–π¬ª) - floors_type;
- —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ –≤ –∫–∏–ª–æ–º–µ—Ç—Ä–∞—Ö - cityCenters_nearest_km.

```python
data['price_m2'] = data['last_price'] / data['total_area']

data['num_day'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
data['num_month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['num_year'] = pd.DatetimeIndex(data['first_day_exposition']).year

data['city_centers_nearest_km'] = data['cityCenters_nearest'] / 1000
data['city_centers_nearest_km'] = data['city_centers_nearest_km'].fillna(100000)
data['city_centers_nearest_km'] = data['city_centers_nearest_km'].astype('int')

def type_floors(data):
    if data['floor'] == 1:
        return '–ø–µ—Ä–≤—ã–π'
    if data['floor'] == data['floors_total']:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    return '–¥—Ä—É–≥–æ–π'
data['floors_type'] = data.apply(type_floors, axis=1)
display(data.head())
data.info()
```
